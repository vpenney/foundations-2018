### Loop through each `tr` to create a list of dictionaries

Each dictionary must contain

- Operator ID
- Operator name
- Mine name
- State
- Mine type
- Coal or metal
- Status
- Commodity

Create a new dictionary for each row.

> - *Tip: Start with an empty dictionary, then add the keys one at a time like we did during class*
> - *Tip: You might want to save all of the cells in a variable, then use indexes to get the second, third, fourth, etc.*
> - *Tip: I know you already skipped a bunch of rows already, but one of them still might be bad! Which one is it? How can you skip it? You might need to slice out some of the end of your list, too. Use `print` to help you debug, or just look at the page closely.*

```python
#mine_dictionary['row_text'] = [each_row.text]  
#row_text.append(mine_dictionary)
#mine_dictionary
```

```python
rows = []

# for loop to pull out each row by operator
for each_row in operator_rows[1:-1]:
    mine_dictionary = {}
    all_cells = each_row.find_elements_by_tag_name("td")
    
    # Operator ID
    mine_dictionary['operator_ID'] = all_cells[0].text
    
    # Operator name
    mine_dictionary['operator_name'] = all_cells[2].text
    
    # mine name
    mine_dictionary['mine_name'] = all_cells[3].text
    
    # state
    mine_dictionary['state'] = all_cells[1].text
    
    # mine type
    mine_dictionary['mine_type'] = all_cells[4].text
    
    # coal or metal
    mine_dictionary['cm'] = all_cells[5].text
    
    # status
    mine_dictionary['status'] = all_cells[6].text
    
    # commodity
    mine_dictionary['commodity'] = all_cells[7].text
    rows.append(mine_dictionary)
rows
    
```

### Save that to a CSV named `dirt-operators.csv`

```python
import pandas as pd
df = pd.DataFrame(rows)
df
```

```python
df.to_csv("mines.csv", index=False)
```
